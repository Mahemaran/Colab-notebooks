### **Pre Processing**

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader("/content/chatprac.txt")
documents = loader.load()
# print(documents)
```

```python
import textwrap
def wrap_text(text, width=100):
    lines=text.split("\n") #split text into line based new lines
    wrapped_lines = [textwrap.fill(line, width=width) for line in lines] #wrap each line indivitually
    wrapped_text = "\n".join(wrapped_lines)
    return wrapped_text
```

```python
wrap_text(str(documents))
```

### **Text Splitting**

```python
from langchain.text_splitter import CharacterTextSplitter
text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=500)
text_splitter = text_splitter.split_documents(documents)
```

```python
# print((text_splitter))
```

### **Embedding**

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
embeddings = HuggingFaceEmbeddings()
db = FAISS.from_documents(text_splitter, embeddings)
```

```python
query = " maran's wife"
doc =db.similarity_search(query)
```

```python
print(wrap_text(str(doc[0])))
```

**Transformers**

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
# model_name = "EleutherAI/gpt-neo-1.3B"
model_name = "EleutherAI/gpt-neo-125M"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
def get_answer_from_document(query):
    docresult = db.similarity_search(query)
    # Concatenate the question with the document context
    input_text = f"{query}\n{docresult}"
    # Tokenize the input
    inputs = tokenizer(input_text, return_tensors="pt", truncation=True, max_length=512)

    # Generate the response
    outputs = model.generate(
        **inputs,
        max_new_tokens=50,         # Increase max_length for a longer response
        temperature=0.8,        # Control randomness
        num_return_sequences=1,
        do_sample=True,
        eos_token_id=tokenizer.eos_token_id)  # Ensure the model knows when to stop
    # Decode and return the result
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response
# Ask a question from the document
user_question = "Maran Profession?"
response = get_answer_from_document(user_question)
print(f"Answer: {response}")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from functools import lru_cache
# Load model and tokenizer
model_name = "EleutherAI/gpt-neo-1.3B"
# model_name = "EleutherAI/gpt-j-6B"
# model_name = "sentence-transformers/all-MiniLM-L6-v2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Cache function for faster responses
@lru_cache(maxsize=1000)  # Cache up to 1000 input variations
def generate_response(input_text: str):
    inputs = tokenizer(input_text, return_tensors="pt")
    outputs = model.generate(
        **inputs,
        max_length=150,
        temperature=0.8,
        num_return_sequences=1,
        do_sample = False,
        eos_token_id=tokenizer.eos_token_id )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
# Example usage
input_text = "who is jacki chan"
response = generate_response(input_text)  # First call (slow)
print(response)
response_cached = generate_response(input_text)  # Cached (fast)
print("Response fetched from cache:", response_cached)
```
