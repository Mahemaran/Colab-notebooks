**importing dependencies**

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.neighbors  import KNeighborsClassifier
```

Read the file and convert to Dataframe

```python
weather_data=pd.read_csv("/content/chennai weather combined_dataset.csv")
# weather_data.head()
print(weather_data.shape)
weather_data.columns
```

```python
weather_data.describe()
```

Remove the unwanted columns

```python
df0=weather_data[['tempmax', 'tempmin', 'temp', 'feelslikemax', 'feelslikemin', 'feelslike', 'dew', 'humidity', 'precip', 'precipprob', 'precipcover', 'windgust','windspeed', 'winddir', 'sealevelpressure', 'cloudcover', 'visibility','solarradiation', 'solarenergy', 'uvindex', 'severerisk', 'moonphase', 'conditions', 'description', 'icon', 'preciptype']]
```

```python
df0["status"]=df0["preciptype"]
df0=df0.drop(columns="preciptype",axis=1)
```

Data cleaning

```python
df0['status'] = df0['status'].fillna("no rain")
df0.fillna(0, inplace=True)
```

```python
df0 = df0.dropna(subset=["winddir", "sealevelpressure", "cloudcover", "visibility"], axis=0, how="any")
```

Convert the string into integer

```python
from sklearn.preprocessing import LabelEncoder
df0['conditions'] = LabelEncoder().fit_transform(df0['conditions'])
df0['description'] = LabelEncoder().fit_transform(df0['description'])
df0['icon'] = LabelEncoder().fit_transform(df0['icon'])
```

```python
X=df0.drop(columns="status", axis=1)
Y=df0.status
```

```python
from sklearn.preprocessing import LabelEncoder
y=LabelEncoder().fit_transform(Y)
X=np.asarray(X)
```

Training or fitting the model for default

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2, stratify=y, random_state=1)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
models=[LogisticRegression(max_iter=10000), SVC(), DecisionTreeClassifier(), RandomForestClassifier(), KNeighborsClassifier()]
```

```python
def check_default_parameter():
    for model in models:
        clf=model.fit(X_train, y_train)
        x_train_predict=clf.predict(X_train)
        x_test_predict=clf.predict(X_test)
        score=cross_val_score(model, X_train, y_train, cv=5)
        print("model name: ", model)
        print(score.mean())
        # print(accuracy_score(y_train,x_train_predict))
        # # print(accuracy_score(y_test,x_test_predict))
        # print(clf.score(X_train, y_train))
        # print(clf.score(X_test, y_test))
        print("------------------")
```

training the model for hyperparameter

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
models=[LogisticRegression(max_iter=10000), SVC(), DecisionTreeClassifier(), RandomForestClassifier(), KNeighborsClassifier()]
```

```python
model_hyperparameter={
                "LogisticRegression":{
                    "C":[1,5,10,20]
                },
                "SVC":{
                    'C':[1,5,10,20],
                    'kernel':["linear","poly","rbf","sigmoid"]
                },
                "DecisionTreeClassifier":{
                    "splitter":['best']
                },
                "RandomForestClassifier":{
                    "n_estimators":[10,20,50,100]
                },
                "KNeighborsClassifier":{
                    "n_neighbors":[3,5,10,20]
                }
}
```

```python
model_key=list(model_hyperparameter.keys())
model_value=list(model_hyperparameter.values())
# print(model_key)
# print(model_value)
```

```python
model_hyperparameter["LogisticRegression"]
```

```python
params=model_hyperparameter['SVC']
print(params)
```

```python
def Model_selection(list_of_models, hyperparameter_dictionary):
    result=[]
    i=0
    for model in list_of_models:
        key=model_key[i]
        params=hyperparameter_dictionary[key]
        i+=1
        print(model)
        print(params)
        classifier=GridSearchCV(model, params, cv=5)
        classifier.fit(X,y)
        result.append({
            'model used': model,
            'highest score':classifier.best_score_,
            "best hyperparameter":classifier.best_params_
        })
    result_dataframe=pd.DataFrame(result, columns=['model used','highest score', "best hyperparameter"])
    return result_dataframe
```

```python
Model_selection(models, model_hyperparameter)
```

best score is RandomForestClassifier and KNN
