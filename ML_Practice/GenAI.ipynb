# ML

**Tom M. Mitchell provided a widely quoted and more formal definition:-**
A computer program is said to learn from experince E with respect to some class of tasks T and performance measure P if its performance at tasks in T, as measured by P, improves with experiance.

* Discriminative AI - logistic regression, SVM, Decision Trees, random forest, etc...
* Modern AI-deep learning, NLP, computer vision, reinforcement learning etc..
* GenAI - content craetion, Genarative adversarial networks(GANs), variational Autoencoders, etc...

**What is ML Algorithm?**

a set of mathematical instructions or rules that enable computers to learn from and make predictions or decision based on data.

**What is Model?**

it is created by trainig an algorithm on a dataset, allowing it to learn patterns and relationship within the data. once its trained the model can make prediction or decision based on new , unseen data
