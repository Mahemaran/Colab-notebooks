### **ML**

**what is standard scaler:**
 its transforms data to have a mean of 0 and a standard deviation of 1. This is also known as **z-score normalization**.
z=x−μ/σ

x: Original value.
μ: Mean of the dataset.
σ: Standard deviation of the dataset

**Purpose:** Algorithms like  logistic regression, SVM, k-NN, k-means clustering rely on distance metrics (e.g., Euclidean distance) and are sensitive to the scale of features and we can use if the feature magnitute difference is high.
