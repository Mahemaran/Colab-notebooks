```python
import pandas as pd
import openpyxl
from matplotlib import pyplot as plt

df = pd.read_excel("/content/Monthly_expence.xlsx", sheet_name="DEC_24", engine='openpyxl')
grouped_df = df.groupby('Category')['Amount'].sum()
print(grouped_df)
df['Amount'].sum()
```

```python
categories = grouped_df.index
amounts = grouped_df.values

# Now you can create the scatter plot
plt.figure(figsize=(20,7))
plt.scatter(categories, amounts)
plt.xlabel("Category") # Setting the x-axis label
plt.ylabel("Amount")   # Setting the y-axis label
plt.xticks(rotation=90, ha='right') # Rotating x-axis labels for readability
plt.tight_layout() # Adjusting layout to prevent labels from overlapping
for i, amount in enumerate(amounts):
    plt.text(categories[i], amounts[i], f'{amount}', ha='center', va='bottom', fontsize=12, color='black')
plt.show()
```
