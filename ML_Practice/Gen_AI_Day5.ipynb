```python
class Circle:
    pi=3.14
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        return Circle.pi*self.radius*self.radius
    def perimeter(self):
        return 2*Circle.pi*self.radius
radius_input = int(input("Enter the radius: "))
obj = Circle(radius_input)
print(obj.area())
print(obj.perimeter())
```

```python
class Employee:
    def __init__(self):
        self._fname=input("Enter first name:")
        self.lname=input("Enter last name:")

    def display(self):
        print("Name:",self._fname)
        print("Name:",self.lname)

emp=Employee()
emp.display()
print(emp._fname)
print(emp.lname)
```

Single Inheritance

```python
class Employee(object):
    def __init__(self):                              #1
        print("Employee init....!")
        self.__fname = input("Enter first name: ")
        self.__lname = input("Enter last name: ")

    def fullname(self):  # Add 'self' as a parameter
        return self.__fname + " " + self.__lname  # Corrected typo '__fnmae' to '__fname'

    def email(self):
        return self.__fname + self.__lname + "@gmail.com"


class Developer(Employee):                           #2
    def __init__(self):
        print("Developer init....!")
        super().__init__()  # Call the parent's __init__ method
        self.__Domain = input("Enter domain: ")

    def display(self):
        print("Fullname: ", self.fullname())  # Call fullname method on self
        print("Email: ", self.email())  # Call email method on self
        print("Domain: ", self.__Domain)


# Creating the Developer object and displaying details
dev = Developer()
dev.display()
```

Multiple Inheritance

```python
class Base1:
    def __init__(self):
        print("Base1")

class Base2:
    def __init__(self):
        print("entered Base2.....!")
        super().__init__()
        print("Base2")

class Derived(Base2,Base1):
    def __init__(self):
        print("entered Derived.....!")
        super().__init__()
        Base1.__init__(self)
        print("derived")

obj=Derived()
```
