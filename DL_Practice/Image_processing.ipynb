**Resizing the image by Pillow**

```python
from PIL import Image
img = Image.open("/content/medium-redcloud-goku-ultra-instinct-wall-poster-for-room-for-original-imagagedrzjxskyj.jpeg")
img_resize = img.resize((200,200))
```

```python
#saving the file
img_resize.save("Goku.jpg")
```

```python
Image.open("/content/Goku.jpg")
```

**Convert RGB to Grayscale using OPenCv**

```python
import cv2
img = cv2.imread("/content/medium-redcloud-goku-ultra-instinct-wall-poster-for-room-for-original-imagagedrzjxskyj.jpeg")
```

```python
img.shape
```

```python
grayscale = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
grayscale.shape
```

```python
grayscale_plot = plt.imshow(grayscale)
```
